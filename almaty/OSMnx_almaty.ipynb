```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Specify the location for Almaty
place = "Almaty, Kazakhstan"

# Load road data from OpenStreetMap using OSMnx
# The "drive" type limits the download to car-accessible roads.
graph = ox.graph_from_place(place, network_type="drive")

# Convert the road graph to a GeoDataFrame
gdf = ox.graph_to_gdfs(graph, nodes=False)

gdf = gdf.to_crs(epsg=32643)

# Visualize the data using GeoPandas and Matplotlib
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, linewidth=0.5, edgecolor="black")
plt.title("Road Network of Almaty")
plt.show()
```
